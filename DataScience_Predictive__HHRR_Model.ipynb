## Investigación
---



### El Costo de Contratar un Empleado en 2024

En este artículo, se analiza el costo promedio por contratación en el año 2024 y los factores que influyen en este costo.

#### Costo Promedio por Contratación

- **2019:** $4,129

- **2023:** $4,700 (aumento del 14%)

#### Factores que Influyen en el Costo

1. **Tamaño de la Empresa:** Las empresas más grandes pueden distribuir los costos operativos entre un mayor número de contrataciones.
2. **Industria:** Ciertas industrias, como la de la ciberseguridad, la ingeniería o la enfermería, experimentan una escasez de talento, lo que aumenta el costo de la contratación.
3. **Ubicación:** Los costos de contratación varían según la ubicación geográfica.

### Tipos de Costos

#### Costos Directos

- **Reclutamiento:** Publicaciones de ofertas de trabajo, honorarios de agencias de reclutamiento, software de seguimiento de candidatos.
- **Selección:** Pruebas, entrevistas, viajes de los candidatos.
- **Onboarding:** Capacitación, materiales, tiempo de los empleados existentes.






#### Costos Indirectos

- **Tiempo de los Empleados:** El tiempo que los empleados dedican al proceso de contratación, como la revisión de currículums y la realización de entrevistas.
- **Pérdida de Productividad:** La pérdida de productividad que se produce cuando un puesto está vacante.
- **Costos de Rotación:** Los costos asociados con la contratación de un nuevo empleado para reemplazar a uno que se ha ido.


Este análisis proporciona una comprensión más profunda del costo de contratar un empleado en 2024 y los diferentes factores que influyen en este costo. Entender estos aspectos es fundamental para la toma de decisiones estratégicas en la gestión de recursos humanos y la planificación financiera de las empresas.

# TAREA #6: ENTRENAR Y EVALUAR UN CLASIFICADOR UTILIZANDO LA REGRESIÓN LOGÍSTICA

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```
X_train.shape
```

```
X_test.shape
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```
y_pred
```

```
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format(100*accuracy_score(y_pred, y_test)))
```

```
# Resultados en el Conjunto de Testing
cm = confusion_matrix( y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred))
```

# TAREA #7: ENTRENAR Y EVALUR UN CLASIFICADOR UTILIZANDO BOSQUES ALEATORIOS

```
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```
y_pred = model.predict(X_test)
```

```
# Resultados en el Conjunto de Testing
cm = confusion_matrix( y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred))
```

# TAREA #8: ENTRENAR Y EVALUR UN CLASIFICADOR UTILIZANDO DEEP LEARNING

```
import tensorflow as tf
```

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 500, activation='relu', input_shape=(50, )))
model.add(tf.keras.layers.Dense(units = 500, activation='relu'))
model.add(tf.keras.layers.Dense(units = 500, activation='relu'))
model.add(tf.keras.layers.Dense(units = 1, activation='sigmoid'))
```

```
model.summary()
```

```
model.compile(optimizer='Adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```
# oversampler = SMOTE(random_state=0)
# smote_train, smote_target = oversampler.fit_sample(X_train, y_train)
# epochs_hist = model.fit(smote_train, smote_target, epochs = 100, batch_size = 50)
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size=50)
```

```
y_pred = model.predict(X_test)
y_pred
```

```
y_pred = (y_pred > 0.5)
y_pred
```

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'])
plt.title("Función de Pérdidas del Modelo durante el Entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Error de Entrenamiento")
plt.legend(["Error de Entrenamiento"])
```

```
plt.plot(epochs_hist.history['accuracy'])
plt.title("Tasa de Acierto del Modelo durante el Entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Accuracy de Entrenamiento")
plt.legend(["Accuracy de Entrenamiento"])
```

```
# Resultados en el Conjunto de Testing
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```
